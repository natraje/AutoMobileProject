# Problem Statement
##### A Chinese automobile company Geely Auto aspires to enter the US market by setting up their manufacturing unit there and producing cars locally to give competition to their US and European counterparts.  They have contracted an automobile consulting company to understand the factors on which the pricing of cars depends. Specifically, they want to understand the factors affecting the pricing of cars in the American market, since those may be very different from the Chinese market. The company wants to know:
 - Which variables are significant in predicting the price of a car
 - How well those variables describe the price of a car

```python
cs_dict = {}
for company in uniqueCompanies:
    mn=round(carpriceDF[carpriceDF['company']==company].price.mean(),2)
    if mn <= 10000:
        cs_dict[company]='lowtier'
    elif mn < 20000:
        cs_dict[company]='stdtier'
    else:
        cs_dict[company]='luxury'

print(cs_dict)
carpriceDF['company_segment'] = carpriceDF['company'].map(cs_dict)
carpriceDF.drop('company',axis=1,inplace=True)
carpriceDF.head()
```

Transform Categorical variable for LR

```python
# 2. Data Preparation 
twoValCols=['fueltype','aspiration','doornumber','enginelocation']
threeValcols=['drivewheel']
def convertColsByDummies(df, col,dropFirst):
    status=pd.get_dummies(df[col],drop_first=True)
    df=pd.concat([df,status],axis=1)
    df=df.drop(col,axis=1)
    return df;

def convertBinarycolumns():
    map= {'fueltype' : {'gas':1, 'diesel':0},
          'aspiration' :{'std':1, 'turbo':0},
          'doornumber' : {'two':0, 'four':1},
          'enginelocation' : {'front':1, 'rear':0}
         }
    return map;

carpriceDF=carpriceDF.replace(convertBinarycolumns());

for col in twoValCols:
    print(col,':',carpriceDF[col].unique())
```

```python
carpriceDF.head()
```

```python
#carpriceDF=pd.getDummies(carpriceDF)
num_vars = list(carpriceDF.columns[carpriceDF.dtypes == 'object'])
print(num_vars)
# Creating dummy variables
df = pd.get_dummies(carpriceDF)
# Droping 1 dummy variable and Keeping n-1 varaibles for each feature 
df.drop(['symboling_risky',
         'carbody_hatchback',
         'drivewheel_4wd',
         'enginetype_l',
         'cylindernumber_three',
         'fuelsystem_1bbl',
         'company_segment_lowtier'],axis=1,inplace=True)
df.columns
```

```python
df.head()
```

- Test and train 
- Featue scaling

```python
## Train and test 
df_train,df_test=train_test_split(df,train_size=0.7,test_size=0.3,random_state=100)
print(df_train.shape)
print(df_test.shape)

# Scaling the variables
# Coeff will be different when you scale the values
## Min Max scaling is used steps are below
### a. Instantiate an object
scaler=MinMaxScaler()

#Find out the scaler columns
scalerCols=[]
for col in df.columns:
    if df[col].max()>1 :
        scalerCols.append(col)

        print(scalerCols)
### b. Fit the data
#### fit() - will fit on the training set, it will learn max and min values
#### transform() - x-xmin/(xmax-xmin)
#### fit_transform() - this does fits and transform
### Rescale only the numeric not the categorical numerics

df_train[scalerCols]=scaler.fit_transform(df_train[scalerCols])
df_train.describe()
```

Dividing into X and Y sets for the model building

```python
y_train = df_train.pop('price')
X_train = df_train
```

## Building our model
We use a mix approach. Initially using the LinearRegression function from SciKit Learn for its compatibility with RFE and then using the Statmodel for statistics analysis of the model

RFE
Recursive feature elimination

```python

# Running RFE with the output number of the variable equal to 15
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 15)             # running RFE and selecting 15 features best describing the price of cars
rfe = rfe.fit(X_train, y_train)
rfe
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
cols = X_train.columns[rfe.support_]
cols
```

```python
notselcols=X_train.columns[~rfe.support_]
notselcols
```


Building model using statsmodel

```python
## Custom funcion
def fitLRM(X_train,y_train):
    X_train = sm.add_constant(X_train)
    lm = sm.OLS(y_train,X_train).fit() 
    print(lm.summary())
    return lm
```

```python
# Build the model using stats model and RFE
X_train_RFE=X_train[cols]
X_train_RFE=sm.add_constant(X_train_RFE)
# Fitting the model 
lm=fitLRM(X_train_RFE,y_train)
```

```python
# wheelbase has more than P>0.005 value 
X_train1 = X_train_RFE.drop('wheelbase', axis=1)
lm1=fitLRM(X_train1,y_train)
```

```python
# stroke has more than P>0.005 value 
X_train2 = X_train1.drop('stroke', axis=1)
lm2=fitLRM(X_train2,y_train)
```

```python
def getVIF(X_train):
    vif = pd.DataFrame()
    X = X_train
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return(vif)
```
